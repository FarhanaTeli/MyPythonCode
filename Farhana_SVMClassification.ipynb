```python
import pandas as pd # For loading csv file data to numpy array
import numpy as np  # For using data as array
from sklearn.model_selection import train_test_split # For splitting data into train and test sets
from sklearn.preprocessing import StandardScaler # For using to find feature scaling
from sklearn.svm import SVC  # SVC class allows to build a kernel SVM model (linear as well as non-linear)
from sklearn.metrics import confusion_matrix, accuracy_score # For using to find confusion matrix and accuracy score
from matplotlib.colors import ListedColormap
import matplotlib.pyplot as plt  # For plotting graph(x,y)
from datetime import datetime # To show current data and time
print('--------------------Implement Classification Algorithm using Support Vector Machine---------------')
print('Name: Teli Farhana Khatoon')
print('Roll No.21')
print('College name: Vivek College of Commerce')
print('M.sc(I.T.)[Sem 2]')
now = datetime.now()
# dd/mm/YY H:M:S
dt_string = now.strftime("%d/%m/%Y %H:%M:%S")
print("Date and Time =", dt_string)	
#To import dataset
data=pd.read_csv('Social_Network_Ads.csv')
data.head() #To show data
```

```python
# Split Dataset into x and y
x=data.iloc[:, [2,3]].values #Independent variable
y=data.iloc[:, 4].values  #Dependent variable
print('Independent Variables (x)-', x)
print('Dependent Variable(Y)–', y)
# Split the X and Y Dataset into the Training set and Testing set
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = 0)
# Perform Feature Scaling
sc = StandardScaler()
x_train = sc.fit_transform(x_train) # Scale the training data
x_test = sc.transform(x_test)
print('---------------After performing feature scaling---------------')
x_test
```

```python
# Fit SVM to the Training set
classifier = SVC(kernel = 'rbf', random_state = 0)
classifier.fit(x_train, y_train)
# Predict the Test Set Results
y_pred = classifier.predict(x_test)
#Make the Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
print('Confusion_Matrix:-', cm)
# Accuracy
ac=accuracy_score(y_test,y_pred)
print('Accuracy:-', ac)
# Visualise the Test set results
x_set, y_set = x_test, y_test
X1, X2 = np.meshgrid(np.arange(start = x_set[:, 0].min() - 1, stop = x_set[:, 0].max() + 1, step = 0.01),
                     np.arange(start = x_set[:, 1].min() - 1, stop = x_set[:, 1].max() + 1, step = 0.01))
plt.contourf(X1, X2, classifier.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(x_set[y_set == j, 0], x_set[y_set == j, 1],
                c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('SVM (Test set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```
