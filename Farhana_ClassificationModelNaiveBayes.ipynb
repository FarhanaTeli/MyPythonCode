```python
import pandas as pd # For loading csv file data to numpy array
from sklearn import preprocessing # Import LabelEncoder
import numpy as np
from sklearn.naive_bayes import GaussianNB # Import Gaussian Naive Bayes model
from sklearn.model_selection import train_test_split #For using to split datasets into train and test sets
from sklearn.metrics import confusion_matrix #For using to find confusion matrix
from sklearn.metrics import accuracy_score # For using to find accuracy of model
import matplotlib.pyplot as plt # For plotting graph(x,y)
import seaborn as sns
from sklearn.metrics import precision_score # For using to find precision
from sklearn.metrics import recall_score # For using to find recall_score of model
from sklearn.metrics import f1_score # For using to finf f1_score
from datetime import datetime # To show current data and time
print('-------------------------Implement Classification Model using Naive Bayes-------------------------')
print('Name: Teli Farhana Khatoon')
print('Roll No.21')
print('College name: Vivek College of Commerce')
print('M.sc(I.T.)[Sem 2]')
now = datetime.now()
# dd/mm/YY H:M:S
dt_string = now.strftime("%d/%m/%Y %H:%M:%S")
print("Date and Time =", dt_string)	
#To import dataset
#Data
data=pd.read_csv('diabetes.csv')
data.head() #To show data
```

```python
# Separate columns into dependent and independent variables
features=data.drop('Outcome', axis=1)
label=data['Outcome']
print('----------Dependent variables----------')
print(features)
print('----------Independent variable----------')
print(label)
# Split daat
# Split data into training sets and test sets
features_train, features_test, label_train, label_test = train_test_split(features, label, test_size = 0.25, random_state =1)
print('--------------------------Train & Test--------------------------')
print("Number of observations and dimensions in training set:", features_train.shape)
print("Number of observations and dimensions in test set:", features_test.shape)
print("Number of observations in training set:", label_train.shape)
print("Number of observations in test set:", label_test.shape)
# Model Development
# Generate a Naive Bayes model on the training set and perform prediction on the test datasets
model=GaussianNB()
# Train the model using training sets
model_train=model.fit(features_train, label_train)
# Predict the response for test datasets
y_pred=model.predict(features_test)
print('-----------Test Datasets Prediction----------')
print(y_pred)
print('-------------------Performance measures-------------------')
cm=confusion_matrix(label_test, y_pred)
# Visualizing Confusion Matrix using Heatmap
# The backend for matplotlib will be set again to inline
%matplotlib inline  
class_names=[0,1] # name  of classes
fig, ax = plt.subplots()
tick_marks = np.arange(len(class_names))
plt.xticks(tick_marks, class_names)
plt.yticks(tick_marks, class_names)
# create heatmap
sns.heatmap(pd.DataFrame(cm), annot=True, cmap="YlGnBu" ,fmt='g', xticklabels=['no diabetes', 'diabetes'],
            yticklabels=['no diabetes', 'diabetes']) # Plot confusion matrix
ax.xaxis.set_label_position("top")
plt.tight_layout()
plt.title('Confusion matrix', y=1.1)
plt.ylabel('Actual label', fontsize=12)
plt.xlabel('Predicted label', fontsize=12)
print(cm, ": is the confusion matrix")
print(accuracy_score(label_test, y_pred), ": is the accuracy score")
print(precision_score(label_test, y_pred), ": is the precision score")
print(recall_score(label_test, y_pred), ": is the recall score")
print(f1_score(label_test, y_pred), ": is the f1 score")
```
